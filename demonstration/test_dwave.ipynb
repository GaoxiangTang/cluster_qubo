```python
import numpy as np
from dwave_qbsolv import QBSolv

# Example usage:
# Define a QUBO problem as a numpy matrix
qubo_size = 500
Q_matrix = np.random.uniform(-1, 1, (qubo_size, qubo_size))

```

```python
def dwave_qbsolve(Q_matrix, subqubo_size):
    """
    Solve the QUBO problem using a numpy matrix and simulated annealing sampler provided by D-Wave's neal.

    Parameters:
    - Q_matrix (np.array): A numpy array representing the QUBO matrix.
    - subqubo_size (int): The size of the sub-QUBO to be solved at once by the annealer.

    Returns:
    - list of dict: A list of dictionaries representing the solution samples where keys are variable indices,
                    and values are the binary values of each variable in the solution.
    """
    # Convert numpy matrix to the dictionary format expected by QBSolv
    Q = {(i, j): Q_matrix[i, j] for i in range(Q_matrix.shape[0]) for j in range(Q_matrix.shape[1]) if Q_matrix[i, j] != 0}

    
    # Use QBSolv to solve the QUBO problem, using the sampler and specified subQUBO size
    response = QBSolv().sample_qubo(Q, solver_limit=subqubo_size, algorithm=0, verbosity=0)
    
    # Extract the best sample from the solution response
    sample = next(iter(response.samples()))

    # Convert the sample dictionary to a numpy array
    solution_vector = np.array([sample[i] for i in range(len(sample))])

    return solution_vector, solution_vector @ Q_matrix @ solution_vector



# Call the function with the QUBO matrix and subQUBO size
solution_samples = dwave_qbsolve(Q_matrix, 100)
print("Solution Samples:", solution_samples[1])  # Print the first sample to check the format
```
